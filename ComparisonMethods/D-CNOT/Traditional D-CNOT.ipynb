```python
# This file is to evaluate traditional D-CNOT as a comparision of EE based D-CNOT. 
# We can apply noise to the only entanglement that will cause degradation(decoherence) and then check the result of D-CNOT.
```

```python
# This code cell is to define the circuit of Traditional D-CNOT. You can visualize the circuit by "circuit.draw("mpl")".
from qiskit import QuantumRegister, ClassicalRegister, QuantumCircuit
from numpy import pi
import numpy as np
from qiskit.visualization import plot_histogram
from qiskit_aer import AerSimulator
from qiskit_aer.noise import (
    NoiseModel,
    QuantumError,
    depolarizing_error,
    pauli_error,
    thermal_relaxation_error,
    phase_amplitude_damping_error,
)
from qiskit.transpiler.preset_passmanagers import generate_preset_pass_manager

qreg_q = QuantumRegister(4, 'q')
creg_c0 = ClassicalRegister(1, 'c0')
creg_c1 = ClassicalRegister(1, 'c1')
circuit = QuantumCircuit(qreg_q, creg_c0, creg_c1)

circuit.h(qreg_q[0]) #the state of controller qubit q0(Alice side) 1/sqrt(2)(|0>+|1>)
circuit.h(qreg_q[1]) 
circuit.cx(qreg_q[1], qreg_q[2]) # the state of entanglement q1q2 is 1/sqrt(2)(|00>+|11>)
circuit.cx(qreg_q[0], qreg_q[1])
circuit.measure(qreg_q[1], creg_c0[0])
circuit.x(qreg_q[2]).c_if(creg_c0, 1)
circuit.cx(qreg_q[2], qreg_q[3])
circuit.h(qreg_q[2])
circuit.measure(qreg_q[2], creg_c1[0])
circuit.z(qreg_q[0]).c_if(creg_c1, 1)
circuit.measure_all()
#circuit.draw("mpl")
```

```python
# This code cell is to perform the ideal simulation (without noise).
# Repeatedly exec circuit for 1000 times.
# The result shows the probability of case:00 + case:11 is 100% which means that the control_qubit is totally related to target_qubit. (D-CNOT's success rate is 100%)
# If case:00 + case:11 is not 100% which means control_qubit is not totally related to target_qubit. (D-CNOT's success rate is not 100%.)
# Especially, if case:00 +case:11 is only 50% which means control_qubit is not realated to target_qubit. (D-CNOT's success rate is 0%.)

sim_ideal = AerSimulator()
counts = sim_ideal.run(circuit,shots=1000).result().get_counts()
target_qubits_counts = {'00': 0, '01':0, '10':0, '11': 0}
for outcome, count in counts.items():
    target_qubits = outcome[0]+outcome[3]  # choose q0(outcome[0]) and q4(outcome[3])
    target_qubits_counts[target_qubits] += count
print(target_qubits_counts)
plot_histogram(target_qubits_counts)
```

```python
# This code cell is to simulate realistic cases (with noise).
# You can change parameter of noise to evaluate the performance of D-CNOT.
# For example, if p_depo=1, the only entanglement will be totally decoherent.
p_depo = 0  # depolarizing error probability,0~1
p_gate_x =0 # pauli X error probability in q-gate of single qubit, 0~0.5
p_gate_z = 0 # pauli Z error probability in q-gate of single qubit, 0~0.5
am_lambda = 0 # the amplitude damping error parameter λ in q-gate of single qubit, 0~0.5
pha_gamma = 0 # the amplitude damping error parameter γ in q-gate of single qubit, 0~0.5

# QuantumError objects
depol_noise = depolarizing_error(p_depo, 2)

pauli_error_x = pauli_error([('X',p_gate_x), ('I', 1 - p_gate_x)])
pauli_error_z = pauli_error([('Z',p_gate_z), ('I', 1 - p_gate_z)])
error_gate_x = pauli_error_x.tensor(pauli_error_x)# pauli X error probability in q-gate of double qubit, 0~1
error_gate_z = pauli_error_z.tensor(pauli_error_z) # pauli Z error probability in q-gate of double qubit, 0~1

pa_damping_error = phase_amplitude_damping_error(am_lambda, pha_gamma)
error_gate_pa_damping = pa_damping_error.tensor(pa_damping_error) # the phase amplitude damping error in q-gate of double qubit, 0~1

# Create noisy simulator backend
noise_set = NoiseModel()
noise_set.add_quantum_error(depol_noise, ['cx'], [1,2])
#noise_set.add_quantum_error(error_gate_x, ['cx'], [1,2])
#noise_set.add_quantum_error(error_gate_z, ['cx'], [1,2])
#noise_set.add_quantum_error(error_gate_pa_damping, ['cx'], [1,2])

sim_noise = AerSimulator(noise_model=noise_set)

# Transpile circuit for noisy basis gates
passmanager = generate_preset_pass_manager(optimization_level=3, backend=sim_noise)
circuit_tnoise = passmanager.run(circuit)

# Run and get counts
counts = sim_noise.run(circuit_tnoise,shots=1000).result().get_counts()
target_qubits_counts = {'00': 0, '01':0, '10':0, '11': 0}
for outcome, count in counts.items():
    target_qubits = outcome[3] + outcome[0]  # choose q0(outcome[3]) and q3(outcome[0])
    target_qubits_counts[target_qubits] += count
print(target_qubits_counts)
plot_histogram(target_qubits_counts)
```
